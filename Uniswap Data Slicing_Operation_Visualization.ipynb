# Data Cleaning

## Appropriately adjust all amounts or quantities in all seven data sets to their correct decimal form.
<br>
<font size="+1">
    <ul>
        <li>Recall, to do this you need to divide the amount by ten to the power of the corresponding decimal. </li>
        <br>
    </ul>
</font>


```python
#for burn_mint dataset
np.power(10, burn_mint.loc[:,'TOKEN_0_DECIMALS'])
burn_mint.loc[:,'LOGGED_AMOUNT0'].astype(float) / np.power(10, burn_mint.loc[:,'TOKEN_0_DECIMALS'])
burn_mint.loc[:,'LOGGED_AMOUNT0'] = burn_mint.loc[:,'LOGGED_AMOUNT0'].astype(float)\
/ np.power(10, burn_mint.loc[:,'TOKEN_0_DECIMALS'])
burn_mint.loc[:,'LOGGED_AMOUNT1'] = burn_mint.loc[:,'LOGGED_AMOUNT1'].astype(float)\
/ np.power(10, burn_mint.loc[:,'TOKEN_1_DECIMALS'])
burn_mint.head()
```

```python
#for swaps dataset
swaps.loc[:, 'AMOUNT0_IN'] = swaps.loc[:,'AMOUNT0_IN'].astype(float) /  np.power(10, swaps.loc[:,'TOKEN_0_DECIMALS'])
swaps.loc[:, 'AMOUNT0_OUT'] = swaps.loc[:,'AMOUNT0_OUT'].astype(float) /  np.power(10, swaps.loc[:,'TOKEN_0_DECIMALS'])
swaps.loc[:, 'AMOUNT1_IN'] = swaps.loc[:,'AMOUNT1_IN'].astype(float) /  np.power(10, swaps.loc[:,'TOKEN_1_DECIMALS'])
swaps.loc[:, 'AMOUNT1_OUT'] = swaps.loc[:,'AMOUNT1_OUT'].astype(float) /  np.power(10, swaps.loc[:,'TOKEN_1_DECIMALS'])
swaps.head()
```

```python
#for other datasets
data_pool=[dai_eth, eth_usdt, uni_eth, usdc_eth, wbtc_eth]
for data in data_pool:
  data.loc[:, 'BALANCE'] = data.loc[:,'BALANCE'].astype(float) /  np.power(10, data.loc[:,'CONTRACT_DECIMALS'])
  data.loc[:, 'TOTAL_SUPPLY'] = data.loc[:,'TOTAL_SUPPLY'].astype(float) /  np.power(10, data.loc[:,'CONTRACT_DECIMALS'])
```

## <font color="red">For every token holder pool (there are five), calculate the percentage of wallet addresses which have a positive balance. Create a new data frame summarizing these percentages and create a plot that best displays this information.</font>


```python
positive_balance_pct=[]
pool_name=["dai_eth", "eth_usdt","usdc_eth", "uni_eth", "wbtc_eth"]
for data in data_pool:
  filter=data["BALANCE"] >0
  positive_balance=round(100*(len(data.loc[filter,'BALANCE'])/len(data["BALANCE"])),4)
  positive_balance_pct.append(positive_balance)
pct_token_holder=pd.DataFrame({"pool":pool_name,
                               "positive_balance_pct":positive_balance_pct})
pct_token_holder.set_index("pool")
```

```python
pct_token_holder.plot(kind="bar",x='pool',y='positive_balance_pct',figsize=(8,6),rot=45)
```

## <font color="red">Plot a time series of liquidity actions (burns and mints) and swaps on the same graph. Be sure to label everything clearly. On another graph, plot the log of all the liquidity actions and swaps.</font>


```python
burn_mint.info()
swaps.info()
```

```python
burn_mint['BLOCK_SIGNED_AT']=pd.to_datetime(burn_mint.loc[:, 'BLOCK_SIGNED_AT']).dt.date
burn_mint['BLOCK_SIGNED_AT']
```

```python
swaps['BLOCK_SIGNED_AT']=pd.to_datetime(swaps.loc[:, 'BLOCK_SIGNED_AT']).dt.date
swaps['BLOCK_SIGNED_AT']
```

```python
count_event_by_date=burn_mint.groupby(['BLOCK_SIGNED_AT','EVENT_TYPE']).size()
df1=pd.DataFrame(count_event_by_date,columns=["count_event"]).unstack().fillna(0)
df1
```

```python
count_swaps_by_date=swaps.groupby(['BLOCK_SIGNED_AT']).size()
df2=pd.DataFrame(count_swaps_by_date,columns=["count_swaps"]).fillna(0)
df2
```

```python
df3=df1.merge(df2, on='BLOCK_SIGNED_AT', how='outer').fillna(0)
df3
```

```python
#visualization
df3.plot(kind="line",figsize=(16,8))
```

```python
df4=np.log(df3)
#visualization
df4.plot(kind="line",figsize=(16,8))
```
